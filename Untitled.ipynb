```python
import requests
from bs4 import BeautifulSoup
import pandas
import os 
import re
```

```python
base_url = 'http://www.kacl780.net'
```

```python
print('Program start...')

#URL to the page with the links to all the different scripts in all the different seasons. 
scripts_list_url = 'http://www.kacl780.net/frasier/transcripts/'

#Request to page, pulling the HTML content
transcripts_request = requests.get(scripts_list_url)
transcripts_request

#Run the HTML content from the request through BeautifulSoup
transcripts_soup =  BeautifulSoup(transcripts_request.text, 'html.parser')

#Creates a list and then populates it with all the content of the a tags on the page. 
#these a tags contain all the HREF tags and URLS, we'll clean them up in a sec. 
#THe first thing to do is collapse the list as it has only 4 elements but many elements in those 4 elements.
links = []
for x in transcripts_soup:
    links.append(transcripts_soup.findAll('a'))

#collapse list, this is something I did in the PLOS data, essentially
# it is a for loop in a for lopp which takes all the elements in all 
#the lists and puts them into one list. 
flat_list = []
for i in links:
	for thing in i:
		flat_list.append(thing)

#this bit of code is obscure to me but I wrote it with Nick's help. 
#Apparently it takes all the times the HREF appears and gets it, but
#with out the HREF tag so it is a clean URL at this point.
#But not all of them go to the episodes.
urls = [i['href'] for i in flat_list]


#Creates a regular expression for finding all the URLs which have
#an episode then number in the URL.
episode = re.compile(r'.*episode_[0-9]+')


#This runs the regular expression through the list of clean urls.
temp_episode_urls = []
for i in urls:
    if episode.match(i):
        temp_episode_urls.append(i)
    
episode_urls = set(temp_episode_urls)
```

```python
temp_episode_urls_2 = []
temp_episode_urls_2.append(set(temp_episode_urls))

episode_urls = []
for i in temp_episode_urls_2:
	for thing in i:
		episode_urls.append(thing)
```
